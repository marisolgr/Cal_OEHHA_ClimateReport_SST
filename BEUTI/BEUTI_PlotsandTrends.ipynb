```python
import matplotlib as plt
import xarray as xr
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stat
from sklearn.preprocessing import PolynomialFeatures
from statsmodels.sandbox.regression.predstd import wls_prediction_std
import statsmodels.api as sm
import cartopy.crs as ccrs
from cartopy.mpl.ticker import LongitudeFormatter, LatitudeFormatter
import cartopy.feature as cfeature

import warnings
warnings.simplefilter('ignore') #filter some warning messages
```

```python
ds = xr.open_dataset('BEUTI_monthly.nc')
ds.close()
ds = ds.sel(time=slice('1988-01-01','2021-12-31'), latitude=slice(32,42))
ds
```

```python
years = np.unique(ds.year.values)
beuti = ds.BEUTI.resample(time='12MS').mean()
beuti 
```

```python
def calculate_trend(x,y,iy,fy):
    # first degree polynomial
    polf = PolynomialFeatures(1)
    # select years
    a = (x>=iy) & (x<=fy)
    x = x[a]
    y = y[a]
    
    # get rid of nan years
    a = ~np.isnan(y)
    y2=y[a]
    x2=x[a]
    y2 = y2.reshape(-1, 1)
    x2 = x2.reshape(-1, 1)

    # calculate years
    xp = polf.fit_transform(x2)
    mods2 = sm.OLS(y2,xp).fit()
    trend = mods2.params[1]*10
    p_tr = mods2.pvalues[1]

    return trend, p_tr
```

```python
colors = plt.cm.coolwarm_r(np.linspace(0,1,11))
tre = list()
ptr = list()
mbeuti = beuti.mean(dim='time').values
smb = np.sort(mbeuti)
plt.figure(figsize=(8,6), dpi=120)
for i, lat in enumerate(beuti.latitude.values):
    c = np.where(smb==mbeuti[i])[0][0]
    plt.plot(years,beuti[:,i]-beuti[:,i].mean()+lat*5,'.-',color=colors[c])
    t,p = calculate_trend(years,beuti[:,i].values,years[0],years[-1])
    tre.append(t)
    ptr.append(p)
plt.yticks([*range(32*5,42*5+1,5)],beuti.latitude.values.astype(int))
plt.ylabel('Latitude', fontsize=14)
plt.grid(True,alpha=0.3, zorder=0)
plt.show()
```

```python
df = pd.DataFrame(data=beuti[:,0].values,columns=[str(beuti.latitude[0].values.astype(int))+'N'],index=years.astype(int))
for i,l in enumerate(beuti.latitude.values.astype(int)[1:]):
    df[str(l)+'N'] = beuti[:,i+1].values
df.to_csv('Upwelling_BEUTI_OceanIndicators.csv')
```
