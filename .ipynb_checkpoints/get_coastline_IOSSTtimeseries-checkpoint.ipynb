```python
# Modules
import warnings
warnings.simplefilter('ignore') # filter some warning messages

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import xarray as xr
import seaborn as sns
import datetime as dt
import fsspec
import s3fs
import scipy.stats as stats
# make datasets display nicely
xr.set_options(display_style="html")  

#magic fncts #put static images of your plot embedded in the notebook
%matplotlib inline  
plt.rcParams['figure.figsize'] = 12, 6
%config InlineBackend.figure_format = 'retina'
```

```python
coast = pd.read_csv('coastal_line.csv', index_col=0)
lats = np.unique(coast['Lat'])
coast.head()
```

```python
### Read Sea Surface Data Available in zar

file_location = 's3://'

ikey = fsspec.get_mapper('s3://noaa-oisstv02r1/', anon=False)

ds_sst = xr.open_zarr(ikey,consolidated=True)

ds_sst
```

```python
assts = np.full((38,len(coast)),np.nan)
for ind, row in coast.iterrows():    
    sst = ds_sst['sst'].sel(time = slice('1982-01-01','2019-12-31'),
                                                lat  = row.Lat,
                                                lon  = row.Lon
                                               ).load()
    
    ssta=sst.resample(time='12MS').mean()
    assts[:,ind]=ssta.data
fechas = ssta.time
```

```python
sst.plot()
plt.grid()
```

```python
coastSST = xr.Dataset(data_vars=dict(SST=(['time','point'],assts),
                                    Lon=(['point'],coast['Lon'].values),
                                    Lat=(['point'],coast['Lat'].values)),
                      coords=dict(point=(coast.index.values),
                                  time=([*range(1982,2020)])))
coastSST
```

```python
coastSST.to_netcdf('coastal_annual_OISST_1982-2019.nc')
```
